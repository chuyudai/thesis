```python
import pandas as pd
```

```python
YA = pd.read_csv('final_YA.csv')
```

```python
income = pd.read_csv('income_by_age_CPI_adjusted.csv')
```

```python
income.head()
```

```python
income_merged = pd.merge(YA, income, how='left', left_on='C0000100',right_on='code')
```

```python
income_merged.head()
```

```python
mom = pd.read_csv('../final.csv')
```

```python
merged = pd.merge(income_merged, mom, how='left', left_on='C0000200',right_on='R0000100')
```

```python
merged.head()
```

```python
merged.to_csv('merged_1979_YAincome.csv')
```
