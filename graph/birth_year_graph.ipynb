```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
YA = pd.read_csv('../final_YA.csv')
NLSY79_year = pd.read_csv('../birth_year/birth_year.csv')
```

```python
child = YA['C0005700']
age_79 = NLSY79_year['R0000500'] + 1900
```

```python
new_df = pd.DataFrame({
    'CHILD79': child,
    'ORG79': age_79
})
```

```python
new_df.head()
```

```python
df_melted = new_df.melt(var_name='Panel', value_name='Birth Year')
df_melted.sort_values(by='Birth Year', inplace=True)
```

```python
plt.figure(figsize=(10, 6))  # Increased width, decreased height
sns.boxplot(x='Birth Year', y='Panel', data=df_melted, hue='Panel', palette='Set2')

# Manually set the x-axis limits to reduce the empty space
plt.xlim(1950, 2015)  # Adjust the limits to fit your data

# Add titles and labels
plt.title('Box Plot of Birth Years for Child and Age_79')
plt.xlabel('Birth Year')
plt.ylabel('Category')

# Show the plot
plt.show()
```
